```python
filtered_df = COMPLETE_df[['ALL_RLA00PCTPROF_1718','ALL_MTH00PCTPROF_1718','DP02_0151PE','DP02_0152PE','SAEPOVRAT5_17RV_PT','DP05_0037PE','DP02_0066PE','DP03_0062E','DP04_0134E','DP03_0009PE','DP02_0112PE','DP02_0071PE']]
filtered_df.to_pickle('filtered_df.pkl')
filtered_df.head()
```

```python
filtered_df.rename(columns = {'ALL_RLA00PCTPROF_1718': 'reading_score', 'ALL_MTH00PCTPROF_1718': 'math_score',
                   'DP02_0151PE':'comp_pct', 'DP02_0152PE':'broadband_pct', 'SAEPOVRAT5_17RV_PT':'poverty_ratio',
                   'DP05_0037PE':'white_pct', 'DP02_0066PE':'college_pct','DP03_0062E':'median_income','DP04_0134E':'median_rent',
                   'DP03_0009PE':'unemployment_pct', 'DP02_0112PE':'home_lang_not_eng', 'DP02_0071PE':'disability_pct'},
                inplace=True)
```

```python
filtered_df.head()
```

```python
filtered_df.info()
```

```python
filtered_df['poverty_ratio'] = pd.to_numeric(filtered_df['poverty_ratio'])
filtered_df.info()
```

```python
filtered_df['comp_pct'] = pd.to_numeric(filtered_df['comp_pct'])
filtered_df['broadband_pct'] = pd.to_numeric(filtered_df['broadband_pct'])
filtered_df['white_pct'] = pd.to_numeric(filtered_df['white_pct'])
filtered_df['college_pct'] = pd.to_numeric(filtered_df['college_pct'])
filtered_df['median_income'] = pd.to_numeric(filtered_df['median_income'])
filtered_df['median_rent'] = pd.to_numeric(filtered_df['median_rent'])
filtered_df['unemployment_pct'] = pd.to_numeric(filtered_df['unemployment_pct'])
filtered_df['home_lang_not_eng'] = pd.to_numeric(filtered_df['home_lang_not_eng'])
filtered_df['disability_pct'] = pd.to_numeric(filtered_df['disability_pct'])
```

```python
filtered_df.info()
```

```python
filtered_df.reading_score = filtered_df.reading_score.apply(str)
PS_values = filtered_df[filtered_df['reading_score'] == "PS" ].index
period_values = filtered_df[filtered_df['reading_score'] == "." ].index
LE_values = filtered_df[filtered_df.reading_score.str.contains("LE")].index
GT_values = filtered_df[filtered_df.reading_score.str.contains("GE")].index
GE_values = filtered_df[filtered_df.reading_score.str.contains("GT")].index
LT_values = filtered_df[filtered_df.reading_score.str.contains("LT")].index

filtered_df.drop(PS_values, inplace = True)
filtered_df.drop(period_values, inplace = True)
filtered_df.drop(LT_values, inplace = True)
filtered_df.drop(LE_values, inplace = True)
filtered_df.drop(GT_values, inplace = True)
filtered_df.drop(GE_values, inplace = True)
filtered_df.info()
```

```python
filtered_df.math_score = filtered_df.math_score.apply(str)
PS_values = filtered_df[filtered_df['math_score'] == "PS" ].index
period_values = filtered_df[filtered_df['math_score'] == "." ].index
LE_values = filtered_df[filtered_df.math_score.str.contains("LE")].index
GT_values = filtered_df[filtered_df.math_score.str.contains("GE")].index
GE_values = filtered_df[filtered_df.math_score.str.contains("GT")].index
LT_values = filtered_df[filtered_df.math_score.str.contains("LT")].index

filtered_df.drop(PS_values, inplace = True)
filtered_df.drop(period_values, inplace = True)
filtered_df.drop(LT_values, inplace = True)
filtered_df.drop(LE_values, inplace = True)
filtered_df.drop(GT_values, inplace = True)
filtered_df.drop(GE_values, inplace = True)
filtered_df.info()
```

```python
def score_cleaner(score):
    score = str(score)
    
    if "-" in score:
        x, y = score.split("-")
        score = (int(x) + int(y))/ 2
        return round(int((score)))
    else:
        return int(score)
```

```python
filtered_df.reading_score = filtered_df.reading_score.apply(score_cleaner)
filtered_df.math_score = filtered_df.math_score.apply(score_cleaner)
filtered_df.info()
```

```python
filtered_df['composite_score'] = (filtered_df['reading_score'] + filtered_df['math_score'])/2
filtered_df.info()
```

```python
filtered_df.to_pickle('../data/CLEANED_df.pkl')
```
